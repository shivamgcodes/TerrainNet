```python
from PIL import Image
import os

itr = 0
string = ["creek", "park-1" , "park-2" , "park-8" , "trail" , "trail-3" , "trail-4" , "trail-5" , "trail-6" , "trail-7" , "trail-9" , "trail-10" , "trail-11" ,"trail-12" , "trail-13" , "trail-14" , "trail-15" ,"village"  ]
for s in string:

    directory_original = r"/content/RUGD/RUGD_frames-with-annotations/RUGD_frames-with-annotations/" + s
    directory_annotated = r"/content/RUGD/RUGD_annotations/RUGD_annotations/" + s

    file_or = os.listdir(directory_original)
    file_ann = os.listdir(directory_annotated)
    image_list_original = []
    image_list_annotated = []

    for file in file_or:
        if file.endswith(('.png' ,'.PNG' )):
            image_list_original.append(Image.open(os.path.join(directory_original , file)))

    for file in file_ann:
        if file.endswith(('.png' , '.PNG')):
            image_list_annotated.append(Image.open(os.path.join(directory_annotated , file)))


    for index in range (len(image_list_annotated)):
        print(s + "  "+  str(index))
        image = image_list_annotated[index]
        image_original = image_list_original[index]

        cropped_imgs = []
        labels = []
        width = image.size[0]
        height = image.size[1]
        w = 0
        h = 0
        while (w + 64 < width):
            h = 0
            while (h + 64 < height):
                left = w
                upper = h
                right = w + 64
                bottom = h + 64
                cur = image.crop((left, upper, right, bottom))
                total = 4096
                dict = {}

                for a in range(left, right):
                    for b in range(upper, bottom):
                        clr = image.getpixel((a, b))
                        if clr in dict:
                            dict[clr] = dict[clr] + 1
                        else:
                            dict[clr] = 1

                largest_value = max(dict.values())
                key_for_largest_value = max(dict, key=lambda k: dict[k])

                if (largest_value / total > 0.9):
                   # cropped_imgs.append(cur)
                    cropped_imgs.append(image_original.crop((left , upper , right , bottom)))
                    labels.append(key_for_largest_value)

                h += 64
            w += 64





        tuple_to_color = {(0,0,0) : "void1" ,
                          (108,64,20):"dirt1" ,
                          (255,229,204):"sand1",
                          (0,102,0):"grass1" ,
                          (0,255,0):"tree1",
                          (0,153,153):"pole1",
                          (0,128,255):"water1",
                          (0,0,255):"sky1",
                          (255,255,0):"vehichle1",
                          (255,0,127):"container1",
                          (64,64,64):"asphalt1",
                          (255,128,0):"garvel1",
                          (255,0,0):"building1",
                          (153,76,0):"mulch1",
                          (102,102,0):"rockbed1",
                          (102,0,0):"log1",
                          (0,255,128):"bicycle1",
                          (204,153,255):"person1",
                          (102,0,204):"fence1",
                          (255, 153,204):"bush1",
                          (0,102,102):"sign1",
                          (153, 204, 255):"rock1",
                          (102,255,255):"bridge1",
                          (101,101,11):"concrete1",
                          (114,85,47):"picnictable1"}


        for i in range(len(cropped_imgs)):

            itr+= 1
            string1 =  tuple_to_color[labels[i]];

            if not os.path.exists(string1):
                os.makedirs(string1)
            cropped_imgs[i].save(string1 + "/" + str(itr) + ".png", "PNG")




```
